# Plot style

## Adding Marekers

```
# Plain plot
x = [1, 2, 3, 4]
y = [2, 4, 6, 8]
plt.plot(x, y)
plt.show()
```

```
# To add a marker we use the keyword argument "marker" 
# and pass the marker type.
# All avalible markers https://matplotlib.org/stable/api/markers_api.html
plt.plot(x, y, marker="s")
plt.show()
```

```
# we can change the marker size and color using the keyword arguments markersize and markerfacecolor respectively
# List of colors and their format is avalible in https://matplotlib.org/stable/tutorials/colors/colors.html
# examples: 'r': red, 'g': green, 'b': blue
plt.plot(x, y, marker="o", markersize=20, markerfacecolor="k", markeredgecolor="y")
plt.show()
```

## Line style and color

```
# We can change the line color using the keyword arguemnt "color"
# The list of colors is avalible in https://matplotlib.org/stable/gallery/color/named_colors.html
plt.plot(x, y, color="g")
plt.show()
```

```
# we can also change the line style using the keyword arguemnt "linestyle"
# different line styles are avalible in https://matplotlib.org/stable/gallery/lines_bars_and_markers/linestyles.html
# examples: dashed, dotted
plt.plot(x, y, color="g", linestyle="dashed")
plt.show()
```

#*Create Title and Labels for a Plot*



2023-02-27 23.47.43.jpg

```
# you can add a title, x and y label using plt.title, plt.x_label, plt.y_label respectively. 
# define the x values
x = np.arange(0, 10, 0.1)
# define the function
y = np.exp(x)
# create the plot
plt.plot(x, y)

# you can set the font parameters with the following dict and pass it to each label function through the fontdict argument
labels_font = {'family':'serif','color':'blue','size':20}
# plot title 
plt.title("Exponential Function",  fontdict=labels_font)
# x title
plt.xlabel("Time", fontdict=labels_font)
# y title
plt.ylabel("Growth Rate", fontdict=labels_font)
plt.show()
```

## Adding a label in multiline plot


```
# you can add a title, x and y label using plt.title, plt.x_label, plt.y_label respectively. 
# define the x values
x = np.arange(0, 10, 0.1)
# define the function
y1 = np.cos(x)
y2 = np.sin(x)
# create the plot
plt.plot(x, y1, label="cos(x)" )
plt.plot(x, y2, label="sin(x)" )
# to show the label use plt.legends
plt.legend()
# you can set the font parameters with the following dict and pass it to each label function through the fontdict argument
labels_font = {'family':'serif','color':'blue','size':20}
# plot title 
plt.title("Sinusoids",  fontdict=labels_font)
# x title
plt.xlabel("X Values", fontdict=labels_font)
# y title
plt.ylabel("Functioin Values", fontdict=labels_font)
plt.show()
```

#*Multiple plots in one figure:*
The subplots() function takes three arguments that describes the layout of the figure.

The layout is organized in rows and columns, which are represented by the first and second argument.

The third argument represents the index of the current plot.

## Example 1

```
# Plot 1:
x = np.arange(-2, 5, 0.1)
cos_x = np.cos(x)

plt.subplot(1, 2, 1)  # The figure has 1 row, 2 columns, and this plot is the first plot.
plt.plot(x, cos_x, color="blue", linewidth=2.5)

# Plot 2:
sin_x = np.sin(x)

plt.subplot(1, 2, 2) #the figure has 1 row, 2 columns, and this plot is the second plot.
plt.plot(x, sin_x, color="red",  linewidth=2.5)

plt.show()
```

## Example 2

```
# Plot 1:
x = np.arange(-2, 5, 0.1)
cos_x = np.cos(x)

plt.subplot(2, 1, 1)  # The figure has 2 row, 1 column, and this plot is the first plot.
plt.plot(x, cos_x, color="blue", linewidth=2.5)

# Plot 2:
sin_x = np.sin(x)

plt.subplot(2, 1, 2) #the figure has 2 row, 1 column, and this plot is the second plot.
plt.plot(x, sin_x, color="red",  linewidth=2.5)
plt.show()
```

#Summary

* plot(x-axis values, y-axis values) — plots a simple line graph with x-axis values against y-axis values

* show() — displays the graph
* title(“string”) — set the title of the plot as specified by the string
* xlabel(“string”) — set the label for x-axis as specified by the string
* ylabel(“string”) — set the label for y-axis as specified by the string
* subplot(nrows, ncols, index) — Add a subplot to the current figure
* suptitle(“string”) — It adds a common title to the figure specified by the string
* subplots(nrows, ncols, figsize) — a convenient way to create subplots, in a single call. It returns a tuple of a figure and number of axes.
* bar(categorical variables, values, color) — used to create vertical bar graphs
* legend(loc) — used to make legend of the graph
* hist(values, number of bins) — used to create a histogram
* scatter(x-axis values, y-axis values) — plots a scatter plot with x-axis values against y-axis values
*axes() — adds an axes to the current figure

## Plotting one plot inside another in the same figure

```
import math
x = np.arange(0, math.pi*2, 0.05)
fig=plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # main axes
axes2 = fig.add_axes([0.55, 0.55, 0.3, 0.3]) # inset axes

y = np.sin(x)
axes1.plot(x, y, 'b')

axes2.plot(x,np.cos(x),'r')

axes1.set_title('sine')
axes2.set_title("cosine")

plt.show()
```
